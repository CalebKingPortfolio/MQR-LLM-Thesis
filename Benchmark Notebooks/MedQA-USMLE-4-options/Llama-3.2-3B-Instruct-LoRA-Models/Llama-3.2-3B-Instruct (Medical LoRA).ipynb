**3. Examine Dataset**

```python
# load the task
task_manager = TaskManager()
task_dict = lm_eval.tasks.get_task_dict(["medqa_4options"], task_manager)
task = task_dict["medqa_4options"]

# build the dataset
task.build_all_requests(rank=0, world_size=1)

# print dataset size
test_docs = list(task.test_docs())
print(f"Test set size: {len(test_docs)}")

# print first example prompt
print("\n" + "="*50)
print("EXAMPLE PROMPT:")
print("="*50)
example = test_docs[0]
print(task.doc_to_text(example))
print("\n" + "="*50)
print("CORRECT ANSWER:")
print("="*50)
print(task.doc_to_target(example))
```

**4. Load LoRA Adaptar Llama-3.2-3B-Instruct (Medical LoRA) Convert From Zip**

```python
# unzip the lora adapter
zip_path = "/content/drive/MyDrive/Fine-Tuning-Notebooks/Low-Rank-Adaptation-(LoRA)/Llama-3.2-3B-Instruct/Medical/medical_llama3_lora_adapter.zip"
extract_path = "/content/llama_medical_lora_adapter"

os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzipped!")
print(f"Adapter location: {extract_path}")
print(os.listdir(extract_path))
```

**5. Load Base Model and Apply Adaptar**

```python
# path to the unzipped lora adapter
adapter_path = "/content/llama_medical_lora_adapter/medical_llama3_lora_adapter"

# base model name from hugging face
base_model_name = "meta-llama/Llama-3.2-3B-Instruct"

# loads the tokenizer from the base model
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(base_model_name)

# add padding token if missing
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# loads the base model
print("Loading base model...")
model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    dtype=torch.bfloat16,
    device_map="auto"
)

# loads the lora adapter on top of the base model
print("Loading LoRA adapter...")
model = PeftModel.from_pretrained(model, adapter_path)

# merge and unload for lm-eval
print("Merging LoRA adapter...")
model = model.merge_and_unload()

# save merged model
print("Saving merged model...")
model.save_pretrained("/content/llama_medical_lora")
tokenizer.save_pretrained("/content/llama_medical_lora")

# confirm paths and files
print(f"\nAdapter path:      {adapter_path}")
print(f"Merged model path: /content/llama_medical_lora")
print(f"Files:             {os.listdir('/content/llama_medical_lora')}")
print("Ready for lm-eval!")
```
