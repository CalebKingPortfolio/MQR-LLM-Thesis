**3. Examine Dataset**

```python
# load the task
task_manager = TaskManager()
task_dict = lm_eval.tasks.get_task_dict(["medqa_4options"], task_manager)
task = task_dict["medqa_4options"]

# build the dataset
task.build_all_requests(rank=0, world_size=1)

# print dataset size
test_docs = list(task.test_docs())
print(f"Test set size: {len(test_docs)}")

# print first example prompt
print("\n" + "="*50)
print("EXAMPLE PROMPT:")
print("="*50)
example = test_docs[0]
print(task.doc_to_text(example))
print("\n" + "="*50)
print("CORRECT ANSWER:")
print("="*50)
print(task.doc_to_target(example))
```

**4. Load Full Model Llama-3.2-3B-Instruct (Medical Full) Convert From Zip**

```python
# path to the zipped model in Google Drive
zip_path = "/content/drive/MyDrive/Fine-Tuning-Notebooks/Full-Fine-Tuning/Llama-3.2-3B-Instruct/Medical/medical_llama3_full_model.zip"

# folder to extract the model into
extract_path = "/content/llama_medical_full"

# make the folder if it does not exist
os.makedirs(extract_path, exist_ok=True)

# unzip the model
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# print to check it worked
print("Unzipped!")

# show files in main folder
print(os.listdir(extract_path))

# show actual model files
print(os.listdir(f"{extract_path}/medical_llama3_full_model"))
```
